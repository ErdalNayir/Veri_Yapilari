```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("/content/drive/MyDrive/MATAY Tubitak/Matay_1.csv",sep = ",")
```

```python
data.head(5)
```

```python
# @title name vs product_serial_no

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(data['name'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(data, x='product_serial_no', y='name', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title product_code

from matplotlib import pyplot as plt
data['product_code'].plot(kind='hist', bins=20, title='product_code')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('name').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
data.groupby('value').size().plot(kind='bar', color=sns.palettes.mpl_palette('Dark2'), figsize=(15,5))
```
